### **Using normal Kmeans clustering**

```python
# import necessary libraries
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
```

```python
#create the first copy to use on kmeans clustering
df1 = df.copy()
```

**Code**

```python
scaler = StandardScaler()  # scale the features
df1[df1.columns] = scaler.fit_transform(df1[df1.columns])
```

```python
# Train the deep clustering model
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(df1)

# Predict the cluster labels for each sample
labels = kmeans.predict(df1)

# Add a new column with the predicted cluster labels
df1['State'] = labels
```

### **Using Kmeans and Keras**

**Code**

```python
from keras.layers import Input, Dense
from keras.models import Model
```

```python
df2 = df.copy()
```

```python
# preprocess the data
scaler = StandardScaler()  # create a scaler object
scaled_df = scaler.fit_transform(df2)  # scale the features
```

```python
# split the data into training and test sets
X_train, X_test = train_test_split(scaled_df, test_size=0.2)
```

```python
# define the model
input_layer = Input(shape=(X_train.shape[1],))  # create input layer
encoded = Dense(128, activation='relu')(input_layer)  # create hidden layer
encoded = Dense(64, activation='relu')(encoded)  # create another hidden layer
encoded = Dense(32, activation='relu')(encoded)  # create another hidden layer
decoded = Dense(64, activation='relu')(encoded)  # create a decoding layer
decoded = Dense(128, activation='relu')(decoded)  # create another decoding layer
decoded = Dense(X_train.shape[1], activation='sigmoid')(decoded)  # create output layer
```

```python
# create the autoencoder model
autoencoder = Model(input_layer, decoded)
```

```python
# compile the model
autoencoder.compile(optimizer='adam', loss='mse')
```

```python
# fit the model on the training data
autoencoder.fit(X_train, X_train, epochs=100, batch_size=32, validation_data=(X_test, X_test))
```

```python
# use the model to cluster the data
clusters = autoencoder.predict(scaled_df)
```

```python
#create clusters
kmeans = KMeans(n_clusters=2)
kmeans.fit(clusters)
clusters = kmeans.predict(clusters)
```

```python
df2.insert(1, 'State', clusters)
```

### **Using Kmeans and Tensorflow**

```python
df3 = df.copy() 
```

```python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
```

```python
df3.head()
```

```python
X = StandardScaler().fit_transform(df3) # scale the features
```

```python
# Split the data into training and test sets
X_train, X_test = train_test_split(X, test_size=0.2)
```

```python
# Define the model
input_layer = Input(shape=(X_train.shape[1],))
encoder = Dense(128, activation='relu')(input_layer)
encoder = Dense(64, activation='relu')(encoder)
encoder = Dense(32, activation='relu')(encoder)
decoder = Dense(64, activation='relu')(encoder)
decoder = Dense(128, activation='relu')(decoder)
output_layer = Dense(X_train.shape[1], activation='sigmoid')(decoder)
```

```python
model = Model(input_layer, output_layer)
model.compile(optimizer='adam', loss='mse')
```

```python
# Train the model
model.fit(X_train, X_train, epochs=100, batch_size=64, validation_data=(X_test, X_test))
```

```python
# Evaluate the model
score = model.evaluate(X_test, X_test, verbose=0)
print(f'Test loss: {score}')
```

```python
# Use the model to cluster the data
encoder_model = Model(input_layer, encoder)
encoded_data = encoder_model.predict(X)
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(encoded_data)
clusters = kmeans.predict(encoded_data)
```

```python
df3.insert(1, 'State', clusters)
```
